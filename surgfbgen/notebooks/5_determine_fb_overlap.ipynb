```python
import os
import pandas as pd
```

```python
pred_path = '~/surgery/surgical_fb_generation/SurgFBGen/outputs/iat_predictions/peskavlp-vision+procedure+task+tracks-none=100-num_tracks=15-final.csv'
pred_df = pd.read_csv(pred_path)

pred_df['h'] = pred_df['cvid'].apply(lambda x: x.split('_')[-1].split('-')[0])
pred_df['m'] = pred_df['cvid'].apply(lambda x: x.split('_')[-1].split('-')[1])
pred_df['s'] = pred_df['cvid'].apply(lambda x: x.split('_')[-1].split('-')[2].split('.')[0])
pred_df['secs'] = pred_df['h'].astype(int) * 3600 + pred_df['m'].astype(int) * 60 + pred_df['s'].astype(int)

n = 0
for i in range(len(pred_df)):
    row = pred_df.iloc[i]
    secs = row['secs']

    tmp_df = pred_df.copy()
    tmp_df['secs'] = tmp_df['secs'] - secs
    tmp_df = tmp_df[(tmp_df['secs'] >= 0) & (tmp_df['secs'] < 10)]
    
    if len(tmp_df) > 1:
        n += 1

n / len(pred_df)
```
