```python
import os
import pandas as pd
```

```python
iat_predictions_dir = '../../outputs/iat_predictions'
new_iat_predictions_dir = '../../outputs/iat_predictions_final'

models = ['peskavlp', 'surgvlp', 'hecvl']
inputs = ['vision', 'vision+procedure', 'vision+procedure+task', 'vision+procedure+task+tracks']
num_none = 100
num_tracks = 15

for model in models:
    for input in inputs:
        if 'tracks' in input:
            file = f'{model}-{input}-none={num_none}-num_tracks={num_tracks}-f1_macro.csv'
        else:
            file = f'{model}-{input}-none={num_none}-f1_macro.csv'
        path = os.path.join(iat_predictions_dir, file)
        new_path = os.path.join(new_iat_predictions_dir, f'{model}-{input}.csv')
        df = pd.read_csv(path)
        df.to_csv(new_path, index=False)
```
